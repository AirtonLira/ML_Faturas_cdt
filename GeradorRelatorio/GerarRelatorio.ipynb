## Classe Relatorio
- **dataframes** *(obrigatório)*: Lista de Dicionaros com Descricao e um Dataframe
- **descricao_problema** *(opcional)*: String descrevendo o problema

- **nome_relatorio** *(opcional)*: String com o nome do relatorio


````python
class Relatorio():
    def __init__(self, dataframes=[], descricao_problema=None, graficos=[], nome_relatorio="relatorio"):
        self.dataframes = dataframes
        self.descricao_problema = descricao_problema
        self.graficos = graficos
        self.nome_relatorio = nome_relatorio
        
        self.gerarRelatorioHTML()
      
    def gerarRelatorioHTML(self):
        self.style = """
            <style>
                body {
                    width: 100%;
                    min-height: 100%;
                    font-family: sans-serif;;
                }

                div {
                    width: 100vh;
                    text-align: center;
                    margin: auto auto;
                }
                
                .capa {
                    height: 99vh;
                    display: flex;
                    flex-direction: column;
                    justify-content: space-between;
                    text-align: center;
                }

                img {
                    align-self: center;
                    width: 400px;
                }

                .info {
                    margin: 250px auto;
                    text-align: center;
                    border-top: 2px solid #DDD;
                    border-bottom: 2px solid #DDD;
                }

                .graficos {
                    margin: 100px auto;
                    text-align: center;
                    display: flex;
                    flex-direction: column;
                }
                
                .dataframes{
                    td, th {
                        padding: 5px;
                    }

                    tr {
                        text-align: center;
                    }

                    tr:nth-child(even) {
                        background: #E0E0E0;
                    }

                    tr:hover {
                        background: silver;
                        cursor: pointer;
                     }
                }

                table {
                    font-size: 11pt; 
                    font-family: Arial;
                    border-collapse: collapse; 
                    border: 1px solid silver;
                    margin-bottom: 30px;
                }

                td, th {
                    padding: 5px;
                }
                
                tr {
                    text-align: center;
                }

                tr:nth-child(even) {
                    background: #E0E0E0;
                }

                tr:hover {
                    background: silver;
                    cursor: pointer;
                 }

            </style>"""
        
        self.html = '''
            <!DOCTYPE html>
            <html lang="en">
            <head>
                <title>Relatório conductor</title>
            </head>
            <body>

                {style}

                <div class="capa">
                    <h3>SQUAD DATASCIENCE</h3>
                    <img src="https://www.riverwoodcapital.com/wp-content/uploads/2016/12/Conductor-Logo.png" alt="Conductor logo">
                    <h3>CONDUCTOR TECNOLOGIA</h3>
                    <h3>João Pessoa, {0}</h3>
                </div>

                <div class="info">
                    <h1>Relatório gerado automaticamente</h1>
                </div>

                <div class="descricao">
                    {1}
                </div>

                <div class="graficos">
                    {2}
                </div>

                <div class="dataframes">
                    <h1>Dataframes</h1>
                    {3}
                </div>

            </body>
            </html>
        '''
        data = date.today().strftime("%d/%m/%Y")
        
        graficos = ""
        if len(self.graficos) > 0: 
            graficos += "<h1>Gráficos</h1>"
            for grafico in self.graficos:
                graficos += "<img src=\"{diretorio}\"/>".format(diretorio=grafico)
        
        descricao = ""
        if self.descricao_problema != None:
            descricao = "<h1>Descrição do problema</h1><p align=\"justify\">{}</p>".format(self.descricao_problema)
        
        dataframes = ""
        for dataframe in self.dataframes:
            dataframes += "<h4 align=\"left\">{}</h4>".format(dataframe["Descricao"])
            dataframes += dataframe["DataFrame"].to_html()

        self.html = self.html.format(data, descricao, graficos, dataframes,style=self.style)
        with open("./temp/"+self.nome_relatorio + 'HTML.html', 'w') as f:
          f.write(self.html)
        
        
        print("Relatório html gerado!")
````
